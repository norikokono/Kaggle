```python
import torch
import torchvision.transforms as T
from torchvision.models.detection import fasterrcnn_resnet50_fpn
import numpy as np
import cv2
from PIL import Image, ImageDraw
import requests
from io import BytesIO
import matplotlib.pyplot as plt
from transformers import AutoModelForCausalLM, AutoTokenizer

class WildfireAnalyzer:
    def __init__(self):
        self.device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
        self.load_models()
        
    def load_models(self):
        """Load detection and analysis models"""
        self.detection_model = fasterrcnn_resnet50_fpn(pretrained=True).eval().to(self.device)
        try:
            model_path = "/kaggle/input/gemma-3n/transformers/gemma-3n-e2b-it/2"
            self.tokenizer = AutoTokenizer.from_pretrained(model_path, trust_remote_code=True)
            self.gemma_model = AutoModelForCausalLM.from_pretrained(
                model_path, 
                device_map="auto",
                torch_dtype=torch.float16 if torch.cuda.is_available() else torch.float32
            )
            self.model_loaded = True
        except Exception as e:
            print(f"Warning: Could not load text generation model: {e}")
            self.model_loaded = False

    def load_image(self, image_url):
        """Load image from URL"""
        try:
            response = requests.get(image_url, stream=True, timeout=10)
            response.raise_for_status()
            return Image.open(BytesIO(response.content)).convert("RGB")
        except Exception as e:
            print(f"Error loading image: {e}")
            return None

    def detect_fire(self, image, confidence_threshold=0.2):
        """Enhanced fire detection with better color verification"""
        transform = T.Compose([T.ToTensor()])
        img_tensor = transform(image).unsqueeze(0).to(self.device)
        
        with torch.no_grad():
            predictions = self.detection_model(img_tensor)[0]
        
        # Convert image to numpy for color analysis
        img_np = np.array(image)
        hsv = cv2.cvtColor(img_np, cv2.COLOR_RGB2HSV)
        
        # Define fire color range in HSV (wider range)
        lower_fire1 = np.array([0, 120, 70])    # Bright red/orange
        upper_fire1 = np.array([30, 255, 255])  # Orange
        lower_fire2 = np.array([0, 70, 50])     # Darker red
        upper_fire2 = np.array([20, 255, 200])  # Dark orange
        
        # Create masks for fire colors
        mask1 = cv2.inRange(hsv, lower_fire1, upper_fire1)
        mask2 = cv2.inRange(hsv, lower_fire2, upper_fire2)
        fire_mask = cv2.bitwise_or(mask1, mask2)
        
        # Find contours in the fire mask
        contours, _ = cv2.findContours(fire_mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
        
        fire_regions = []
        for contour in contours:
            if cv2.contourArea(contour) < 50:  # Minimum area threshold
                continue
                
            x, y, w, h = cv2.boundingRect(contour)
            box = [x, y, x + w, y + h]
            region = img_np[y:y+h, x:x+w]
            
            if region.size == 0:
                continue
                
            # Calculate fire ratio and confidence
            region_hsv = cv2.cvtColor(region, cv2.COLOR_RGB2HSV)
            region_fire_mask = cv2.bitwise_or(
                cv2.inRange(region_hsv, lower_fire1, upper_fire1),
                cv2.inRange(region_hsv, lower_fire2, upper_fire2)
            )
            fire_ratio = np.count_nonzero(region_fire_mask) / region_fire_mask.size
            
            if fire_ratio > 0.05:
                intensity = np.mean(region[:,:,0][region_fire_mask > 0]) if np.any(region_fire_mask) else 0
                confidence = min(0.99, fire_ratio * 2.0)
                
                fire_regions.append({
                    'box': [int(coord) for coord in box],
                    'confidence': float(confidence),
                    'intensity': float(intensity),
                    'area': w * h
                })
        
        return fire_regions

    def detect_smoke(self, image, min_coverage=0.05):
        """Enhanced smoke detection with better filtering"""
        img_np = np.array(image)
        hsv = cv2.cvtColor(img_np, cv2.COLOR_RGB2HSV)
        
        # Define smoke color ranges with better thresholds
        lower_smoke1 = np.array([0, 0, 150])    # Bright smoke
        upper_smoke1 = np.array([180, 50, 255])
        lower_smoke2 = np.array([0, 0, 100])    # Darker smoke
        upper_smoke2 = np.array([180, 30, 200])
        
        # Create masks for smoke colors
        mask1 = cv2.inRange(hsv, lower_smoke1, upper_smoke1)
        mask2 = cv2.inRange(hsv, lower_smoke2, upper_smoke2)
        smoke_mask = cv2.bitwise_or(mask1, mask2)
        
        # Apply morphological operations
        kernel = np.ones((5,5), np.uint8)
        smoke_mask = cv2.morphologyEx(smoke_mask, cv2.MORPH_OPEN, kernel)
        smoke_mask = cv2.morphologyEx(smoke_mask, cv2.MORPH_CLOSE, kernel)
        
        # Calculate smoke coverage
        smoke_coverage = (np.count_nonzero(smoke_mask) / smoke_mask.size) * 100
        
        # Only return significant smoke detections
        if smoke_coverage < min_coverage:
            return np.zeros_like(smoke_mask), 0.0
            
        return smoke_mask, smoke_coverage

    def generate_visualization(self, image, fire_regions, smoke_mask):
        """Generate visualization of detections"""
        vis_image = image.copy()
        draw = ImageDraw.Draw(vis_image, 'RGBA')
        
        # Draw smoke regions (semi-transparent blue)
        if smoke_mask is not None and np.any(smoke_mask > 0):
            smoke_vis = Image.fromarray((smoke_mask * 0.5).astype(np.uint8), 'L')
            smoke_color = Image.new('RGBA', image.size, (0, 0, 255, 64))
            vis_image.paste(smoke_color, (0, 0), mask=smoke_vis)
        
        # Draw fire regions (red bounding boxes)
        for region in fire_regions:
            box = region['box']
            confidence = region['confidence']
            
            # Draw rectangle
            draw.rectangle(box, outline='red', width=3)
            
            # Add label
            label = f"Fire {confidence:.0%}"
            text_bbox = draw.textbbox((box[0], box[1] - 20), label)
            draw.rectangle(text_bbox, fill='red')
            draw.text((box[0], box[1] - 20), label, fill='white')
        
        return vis_image

    def generate_analysis(self, fire_regions, smoke_coverage, visualization_image):
        """Generate analysis based on detections"""
        try:
            # Calculate fire statistics
            total_fire_area = sum(r['area'] for r in fire_regions) if fire_regions else 0
            fire_detected = len(fire_regions) > 0
            smoke_detected = smoke_coverage > 1.0  # At least 1% coverage
            
            # Create analysis
            analysis = "ğŸ”¥ WILDFIRE DETECTION ANALYSIS\n"
            analysis += "=" * 50 + "\n\n"
            
            # Fire analysis
            if fire_detected:
                analysis += "ğŸ”¥ FIRE DETECTED\n"
                analysis += "-" * 30 + "\n"
                analysis += f"â€¢ Number of fire regions: {len(fire_regions)}\n"
                analysis += f"â€¢ Total fire area: {total_fire_area:,} pixels\n"
                
                # Calculate average intensity
                avg_intensity = sum(r['intensity'] for r in fire_regions)/len(fire_regions)
                analysis += f"â€¢ Average fire intensity: {avg_intensity:.1f}/255\n"
                
                # Classify fire severity
                if total_fire_area > 5000:  # pixels
                    severity = "ğŸ”´ HIGH - Large fire detected"
                elif total_fire_area > 1000:
                    severity = "ğŸŸ  MEDIUM - Moderate fire detected"
                else:
                    severity = "ğŸŸ¡ LOW - Small fire detected"
                analysis += f"â€¢ Severity: {severity}\n"
            else:
                analysis += "â„¹ï¸ No significant fire detected\n"
            
            # Smoke analysis
            analysis += "\nğŸ’¨ SMOKE ANALYSIS\n"
            analysis += "-" * 30 + "\n"
            if smoke_detected:
                analysis += f"â€¢ Smoke coverage: {smoke_coverage:.1f}% of image\n"
                
                # Classify smoke severity
                if smoke_coverage > 20:
                    severity = "ğŸ”´ HIGH - Heavy smoke detected"
                elif smoke_coverage > 5:
                    severity = "ğŸŸ  MEDIUM - Moderate smoke detected"
                else:
                    severity = "ğŸŸ¡ LOW - Light smoke detected"
                analysis += f"â€¢ Severity: {severity}\n"
            else:
                analysis += "â€¢ No significant smoke detected\n"
            
            # Recommendations
            analysis += "\nğŸš¨ RECOMMENDED ACTIONS\n"
            analysis += "-" * 30 + "\n"
            if fire_detected or smoke_detected:
                if fire_detected:
                    analysis += "â€¢ ğŸ”¥ Evacuate the area if fire is nearby\n"
                    analysis += "â€¢ ğŸš’ Alert local fire department immediately\n"
                if smoke_detected:
                    analysis += "â€¢ ğŸ˜· Wear a mask if going outside\n"
                    analysis += "â€¢ ğŸ  Stay indoors and close windows if smoke is heavy\n"
                analysis += "â€¢ ğŸ“¡ Monitor local emergency broadcasts\n"
            else:
                analysis += "â€¢ No immediate action required\n"
                analysis += "â€¢ Continue to monitor the situation\n"
            
            return analysis
            
        except Exception as e:
            print(f"Error in analysis: {e}")
            return "Error generating analysis. Please check the input image and try again."

    def analyze_image(self, image_url):
        """Main analysis pipeline"""
        try:
            # Load image
            image = self.load_image(image_url)
            if image is None:
                return {"error": "Failed to load image"}
            
            # Detect fire and smoke
            fire_regions = self.detect_fire(image)
            smoke_mask, smoke_coverage = self.detect_smoke(image)
            
            # Generate visualization
            vis_image = self.generate_visualization(image, fire_regions, smoke_mask)
            
            # Generate analysis
            analysis = self.generate_analysis(fire_regions, smoke_coverage, vis_image)
            
            return {
                "original_image": image,
                "visualization": vis_image,
                "fire_regions": fire_regions,
                "smoke_coverage": smoke_coverage,
                "analysis": analysis
            }
            
        except Exception as e:
            return {"error": f"Error in analysis: {str(e)}"}

def main():
    print("Initializing Wildfire Analyzer...")
    analyzer = WildfireAnalyzer()
    
    # Example image URL
    image_url = "https://www.usfa.fema.gov/img/photos/satellite-wildfire-image-1800w.webp"
    
    print(f"Analyzing image: {image_url}")
    result = analyzer.analyze_image(image_url)
    
    # Display results
    if "error" in result:
        print(f"Error: {result['error']}")
    else:
        print("\n" + "="*70)
        print("WILDFIRE DETECTION RESULTS")
        print("="*70)
        print(f"Fire regions detected: {len(result['fire_regions'])}")
        if result['fire_regions']:
            avg_confidence = sum(r['confidence'] for r in result['fire_regions'])/len(result['fire_regions'])
            print(f"Average fire confidence: {avg_confidence:.1%}")
        print(f"Smoke coverage: {result['smoke_coverage']:.1f}%")
        
        print("\n" + "="*70)
        print("ANALYSIS")
        print("="*70)
        print(result["analysis"])
        
        # Display images
        plt.figure(figsize=(20, 10))
        plt.subplot(1, 2, 1)
        plt.imshow(result["original_image"])
        plt.title("Original Image")
        plt.axis('off')
        
        plt.subplot(1, 2, 2)
        plt.imshow(result["visualization"])
        plt.title("Fire (Red) and Smoke (Blue) Detection")
        plt.axis('off')
        
        plt.tight_layout()
        plt.show()

if __name__ == "__main__":
    main()
```
